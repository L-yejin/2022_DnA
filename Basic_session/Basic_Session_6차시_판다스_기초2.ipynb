````python
# 사용할 함수 정의
def tenFunc(x):
    return x*10
````

#### 

### <span style="color:gray">map</span>  

series 인식 -> series 리턴

````python
# 데이터 프레임 적용시 에러 발생

#abalone.map(tenFunc)

abalone[['length']].map(tenFunc)
````

````python
# 하나의 column에만 적용

abalone['length'].map(tenFunc)
````

````python
# row에 적용

abalone.iloc[1].map(tenFunc)
````

#### 

### <span style="color:gray">apply</span>  

dataframe, series 입력 -> dataframe, series 리턴  

````python
# 전체 데이터프레임 -> 데이터프레임

abalone.apply(tenFunc)
````

````python
# 데이터프레임 -> 데이터프레임

abalone[['length']].apply(tenFunc)
````

````python
# 시리즈 -> 시리즈

abalone['length'].apply(tenFunc)
````

````python
# row에 적용

abalone.iloc[1].apply(tenFunc)
````

#### 

### <span style="color:gray">applymap</span>  

dataframe 입력 -> dataframe 리턴

````python
# 전체 데이터에 적용

abalone.applymap(tenFunc)
````

````python
# 데이터 프레임 -> 데이터 프레임

abalone[['length']].applymap(tenFunc)
````

````python
# 시리즈 적용시 에러 발생

#abalone['length'].applymap(tenFunc)

# row에 적용
# 이것도 시리즈여서 에러 발생
abalone.iloc[1].applymap(tenFunc)
````

### <span style="color:gray">-split</span>

````python
# 컬럼sex의 종류 살펴보기
abalone['sex'].value_counts()
````

````python
# 전체 데이터의 5번째까지 컬럼만 사용해서 키별로 데이터 뽑기

for key,group in abalone.iloc[:,:5].groupby(abalone['sex']):
    print('key: ', key)
    print('number: ', len(group))
    print(group)
````

````python
# get_group()
# sex에서 F값만 출력
# 위의 코드에서 key:F 부분과 같은내용

abalone.groupby('sex').get_group('M')
````

#### 

### pivot table

데이터의 **요약된 정보**를 출력  

<img src='http://drive.google.com/uc?export=view&id=1TBrfboGZ5V2vftrhSk3S1VdSD-nyKoED' /><br>

```python
pd.pivot_table(df,                 # 데이터 프레임
               values=None,       # 분석할 열
               index=None,        # index로 사용될 컬럼
               columns=None,      # column으로 사용될 컬럼
               aggfunc='mean',     # 집계함수(데이터 축약시 사용할 함수)
               fill_value=None,   # NaN값 대체할 값
               margins=False,     # 분석 결과를 오른쪽과 아래에 붙일지 여부(행/열 별 총합)
               dropna=True,       # 
               margins_name='All') # 마진 열(행)의 이름
```

````python
df = pd.DataFrame({'A':[2000,1988,2000,2000,1998,1996,1996],
                  'B':[3,12,12,3,12,3,12],
                  'C':[20,1,20,1,20,1,1],
                  'D':[3,3,5,0,5,0,0],
                  'E':[1,2,1,2,2,1,1]})

df
````

````python
pd.pivot_table(df,
               #values='C', 
               index='A', 
               columns='B', 
               aggfunc='mean', 
               fill_value=None, 
               margins=False, 
               dropna=True, 
               margins_name='All')
````

````python
pd.pivot_table(df,
               values=['C','D','E'], 
               index='A', 
               columns='B', 
               aggfunc='mean', 
               fill_value=None, 
               margins=False, 
               dropna=True, 
               margins_name='All')
````

````python
pd.pivot_table(df,
               values='C', 
               index='A', 
               columns='B', 
               aggfunc='mean', 
               fill_value=None, 
               margins=False, 
               dropna=True, 
               margins_name='All')
````

````python
df.pivot_table(values='C', 
               index='A', 
               columns='B', 
               aggfunc='mean', 
               fill_value=None, 
               margins=False, 
               dropna=True, 
               margins_name='All')
````

````python
pd.pivot_table(df,
               values='C', 
               index='A', 
               columns='B', 
               aggfunc='sum', 
               fill_value=0, 
               margins=True, 
               dropna=True, 
               margins_name='All_sum')
````

# Crosstab

범주형 변수를 기준으로 개수 파악이나 수치형 데이터를 넣어 계산할 때 사용  

<img src='http://drive.google.com/uc?export=view&id=1f4x5TaUNQawRQwAiPRt6V5KguCEvTPH0' /><br>

```python
pd.crosstab(index,
            columns,
            values,
            aggfunc,
            rownames,
            colnames,
            margins,
            normalize)
```

````python
df = pd.DataFrame([
        ['ID01', 2, 139, 'country', 'no', 3],
        ['ID02', 3, 148, 'country', 'no', 5],
        ['ID03', 3, 149, 'country', 'no', 7],
        ['ID04', 5, 151, 'country', 'no', 10],
        ['ID05', 7, 154, 'city', 'no', 12],
        ['ID06', 2, 149, 'country', 'no', 7],
        ['ID07', 8, 155, 'city', 'yes', 13],
        ['ID08', 9, 155, 'city', 'yes', 13],
        ['ID09', 6, 154, 'city', 'no', 12],
        ['ID10', 9, 156, 'city', 'yes', 13],
        ['ID11', 6, 153, 'city', 'no', 12],
        ['ID12', 2, 151, 'country', 'no', 6]
    ], columns=['ID', 'hour', 'A', 'B', 'C', 'score'])

df
````

````python
pd.crosstab(df, index='B', columns='C')
````

````python
df.crosstab(index='B',
            columns='C')
````

````python
pd.crosstab(index=df['B'],
            columns=df['C'])
````

````python
pd.crosstab(values=df['score'],
            index=df['B'],
            columns=df['C'],
            aggfunc='mean')
````
